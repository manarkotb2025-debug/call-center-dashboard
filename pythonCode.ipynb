```python
import matplotlib.pyplot as plt
import seaborn as sns

actual_calls = Calls.groupby('Date').size().reset_index(name='Actual Calls')
actual_calls['Date'] = pd.to_datetime(actual_calls['Date'])

forecast_daily = Forecast.groupby('Date')['Forecast Calls'].sum().reset_index()

comparison = pd.merge(actual_calls, forecast_daily, on='Date')

plt.figure(figsize=(12, 6))
plt.plot(comparison['Date'], comparison['Actual Calls'], label='Actual', color='blue')
plt.plot(comparison['Date'], comparison['Forecast Calls'], label='Forecast', color='orange', linestyle='--')
plt.title('Actual Calls vs Forecasted Calls')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=Surveys, x='Satisfaction Score', palette='viridis')
plt.title('Distribution of Customer Satisfaction Scores')
plt.xlabel('Score')
plt.ylabel('Number of Surveys')
plt.show()
```

```python
survey_dept = Surveys.merge(Agents, on='Agent ID').merge(Departments, on='Department ID')

res_rate = survey_dept.groupby('Department Name')['Resolution'].mean() * 100

plt.figure(figsize=(10, 6))
res_rate.sort_values().plot(kind='barh', color='skyblue')
plt.title('Resolution Rate by Department (%)')
plt.xlabel('Percentage')
plt.show()
```

```python
Calls['Hour'] = pd.to_datetime(Calls['Time']).dt.hour
Calls['Day'] = pd.to_datetime(Calls['Date']).dt.day_name()

heatmap_data = Calls.pivot_table(index='Day', columns='Hour', values='Agent ID', aggfunc='count')

plt.figure(figsize=(15, 8))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=False)
plt.title('Call Volume Heatmap (Day vs Hour)')
plt.show()
```
